**827** (861) days available 
- from **2019-08-05** to **2019-10-16** (73d)
- from **2019-11-20** to **2021-12-12** (754d)

### * Two steps to recover the data

(1) create a list/dataframe with entire 79,392 timestamps ('%Y-%m-%d %H:%M:%S')<br>
(2) extract the timestamps ('%Y-%m-%d %H:%M:%S') from the raw data<br>
(3) compare the two list/dataframe and do the interpolation for the raw data<br>
(4) the new data should be the length of 79,392

## 1. Calculate the average and sum traffic flow

#### &nbsp;&nbsp;&nbsp;&nbsp; Select the sites with highest traffic flow

```python
detectors = pd.read_csv('../../../Data/Data processing/20211028_SCOOT_Data/JupyterNotebook_CSV/useful_detector_530.csv')
detectors.sort_values(by=['zeroPercentage%'])
```

```python
means = []
sums = []
zeroPercent = []
errorNum = []
consecutiveError = []
sites = []

for i in detectors.siteId.tolist():
    print(i)
    site_df = pd.read_csv('../../../Data/Data processing/20211028_SCOOT_Data/JupyterNotebook_CSV/20211213_SCOOT_data/' + i + '&2019_08_05-2021_12_12.csv')
    average_flow = site_df['flow'].mean()
    sum_flow = site_df['flow'].sum()
    zero = detectors.loc[detectors.siteId == i, 'zeroPercentage%'].item()
    error = 79392 - len(site_df)
    error2 = sorted(list(site_df.loc[site_df['count'] > 2, 'count'].unique()), reverse=True)    
    
    means.append(average_flow)
    sums.append(sum_flow)
    zeroPercent.append(zero)
    errorNum.append(error)
    consecutiveError.append(error2)
    sites.append(i)
    
    
flow_stats = {'siteId': sites,
              'averageFlow': means,
              'sumFlow': sums,
              'zeroPercentage(%)': zeroPercent,
              'errorNum': errorNum,
              'consecutiveError': consecutiveError}

flow_stats_df = pd.DataFrame(flow_stats)
```

```python
#merge the coordinates
flow_stats_df = pd.merge(flow_stats_df, detectors[['siteId', 'lat', 'lon', 'easting', 'northing']], on="siteId")
```

```python
from datetime import datetime, timedelta

def get_time_string_list(start,end,interval):
    start = datetime.strptime(start, "%Y-%m-%d %H:%M:%S")
    end = datetime.strptime(end, "%Y-%m-%d %H:%M:%S")
    now = start
    string_list = []
    while now <= end:
        now += timedelta(minutes=15)
        string_list.append(now.strftime("%Y-%m-%d %H:%M:%S"))
    return string_list

time_list = get_time_string_list("2019-08-04 23:45:00","2021-12-12 23:30:00",15)
```
