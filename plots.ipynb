```python
def make_plot(data,model):
    # Creating DataFrame
    df = pd.DataFrame(data)

    # Plotting
    plt.figure(figsize=(14, 10), dpi=300)

    # Main title for the figure
    plt.suptitle(model+' Metric Analysis vs Input Length', fontsize=16, fontweight='bold')

    # Plot for MSE_val
    plt.subplot(2, 2, 1)
    plt.plot(df['input_len'], df['MSE_val'], marker='o', linestyle='-', label='MSE_val')
    plt.xlabel('Input Length')
    plt.ylabel('MSE_val')
    plt.title('MSE_val vs Input Length')
    plt.xticks(df['input_len'], rotation=45) # change the xsticks to match the input_len
    plt.grid(True, color='lightgrey', linewidth=0.7)
    plt.legend()

    # Plot for MAE_val
    plt.subplot(2, 2, 2)
    plt.plot(df['input_len'], df['MAE_val'], marker='o', linestyle='-', label='MAE_val', color='orange')
    plt.xlabel('Input Length')
    plt.ylabel('MAE_val')
    plt.title('MAE_val vs Input Length')
    plt.xticks(df['input_len'], rotation=45) # change the xsticks to match the input_len
    plt.grid(True, color='lightgrey', linewidth=0.7)
    plt.legend()

    # Plot for RMSE_val
    plt.subplot(2, 2, 3)
    plt.plot(df['input_len'], df['RMSE_val'], marker='o', linestyle='-', label='RMSE_val', color='green')
    plt.xlabel('Input Length')
    plt.ylabel('RMSE_val')
    plt.title('RMSE_val vs Input Length')
    plt.xticks(df['input_len'], rotation=45) # change the xsticks to match the input_len
    plt.grid(True, color='lightgrey', linewidth=0.7)
    plt.legend()

    # Plot for MAPE_val
    plt.subplot(2, 2, 4)
    plt.plot(df['input_len'], df['MAPE_val'], marker='o', linestyle='-', label='MAPE_val', color='red')
    plt.xlabel('Input Length')
    plt.ylabel('MAPE_val (%)')
    plt.title('MAPE_val vs Input Length')
    plt.xticks(df['input_len'], rotation=45) # change the xsticks to match the input_len
    plt.grid(True, color='lightgrey', linewidth=0.7)
    plt.legend()

    plt.tight_layout()
    #plt.savefig("figures/line_plot_cnn.png", bbox_inches='tight',dpi=300)
    plt.show()
```
