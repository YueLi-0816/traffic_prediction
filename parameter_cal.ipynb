# 2. Calculate the number of parameters 

## 2.1 CNN

```python
def calculate_cnn_params(hyperparams, input_length):
    filters = hyperparams['filters']
    kernel_size = hyperparams['kernel_size']
    dropout = hyperparams['dropout']  # Not used for parameter calculation
    output_neurons = 6

    # First Conv1D Layer
    first_conv_params = (kernel_size * 1 + 1) * filters

    # Second Conv1D Layer
    second_conv_params = (kernel_size * filters + 1) * filters

    # Flatten Layer: Compute input size after pooling
    pooled_length = (input_length - 2 * (kernel_size - 1)) // 2  # Assuming valid padding
    flattened_size = pooled_length * filters

    # Dense Layer
    dense_params = (flattened_size + 1) * output_neurons

    # Total Parameters
    total_params = first_conv_params + second_conv_params + dense_params
    return total_params
```

```python
# Calculate parameters for each input length
results_cnn = []
for length, hyperparams in best_hyperparameters.items():
    total_params = calculate_cnn_params(hyperparams, length)
    results_cnn.append({'Input Length': length, 'Total Params (CNN-Dataset After COVID)': total_params})

# Create a dataframe for the results
df_cnn = pd.DataFrame(results_cnn)
df_cnn['Total Params (CNN-Dataset After COVID)'] = df_cnn['Total Params (CNN-Dataset After COVID)']#.apply(lambda x: '{:,.0f}'.format(x))
df_cnn.sort_values(by='Input Length', inplace=True)
```

```python
# Calculate parameters for each input length
results_cnn_all = []
for length, hyperparams in best_hyperparameters_cnn.items():
    total_params = calculate_cnn_params(hyperparams, length)
    results_cnn_all.append({'Input Length': length, 'Total Params (CNN-Entire Dataset)': total_params})

# Create a dataframe for the results
df_cnn_all = pd.DataFrame(results_cnn_all)
df_cnn_all['Total Params (CNN-Entire Dataset)'] = df_cnn_all['Total Params (CNN-Entire Dataset)']#.apply(lambda x: '{:,.0f}'.format(x))
df_cnn_all.sort_values(by='Input Length', inplace=True)
```

## 2.2 LSTM

```python
def calculate_lstm_params(hyperparams, input_length):
    units = hyperparams['units']
    output_neurons = 6  # Output layer size
    
    # First LSTM Layer
    first_lstm_params = 4 * ((1 + units) * units + units)  # input_size = 1 (from input_shape)

    # Second LSTM Layer
    second_lstm_params = 4 * ((units + units) * units + units)  # input_size = units

    # Dense Layer
    dense_params = (units + 1) * output_neurons

    # Total Parameters
    total_params = first_lstm_params + second_lstm_params + dense_params
    return total_params
```
